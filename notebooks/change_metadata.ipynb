```python
import pandas as pd
import numpy as np

def get_AID2affiliation(paper_affiliations, paper2pubyear_dict):
    # Drop rows where 'Institution' is NaN or its length exceeds 85
    valid_affiliations = paper_affiliations.dropna(subset=['Institution'])
    valid_affiliations = valid_affiliations[(valid_affiliations['Institution'].str.len() <= 85) & (valid_affiliations['Institution'].str.len() > 4)]
    
    # Add a new column 'pubyear' from the paper2pubyear_dict
    valid_affiliations['pubyear'] = valid_affiliations['PMID'].map(paper2pubyear_dict)
    
    # Drop rows where 'pubyear' is NaN (i.e., where PMID wasn't found in paper2pubyear_dict)
    valid_affiliations = valid_affiliations.dropna(subset=['pubyear'])

    # Convert 'pubyear' to integer (in case it's not)
    valid_affiliations['pubyear'] = valid_affiliations['pubyear'].astype(int)
    
    # For each AID, keep the row with the latest 'pubyear'
    latest_affiliations = valid_affiliations.loc[valid_affiliations.groupby('AID')['pubyear'].idxmax()]

    # Create the AID2affiliation dictionary
    AID2affiliation = latest_affiliations.set_index('AID')['Institution'].to_dict()

    return AID2affiliation

AID2affiliation = get_AID2affiliation(paper_affiliations_10k, paper2pubyear_dict)
```

```python
import numpy as np
# Define the file paths
file_path1 = '/data/jx4237data/CM4AI data/dictionary_NIHbmic_databases_deduplicates.csv'
file_path2 = '/data/jx4237data/CM4AI data/dictionary_EBI_databases_deduplicates.csv'
file_path3 = '/data/jx4237data/CM4AI data/dictionary_NAR_databases_deduplicate.csv'

# Read the CSV files
NIHbmic_databases_deduplicates = pd.read_csv(file_path1)
EBI_databases_deduplicates = pd.read_csv(file_path2)
NAR_databases_deduplicate = pd.read_csv(file_path3)
NAR_databases_deduplicate['AltName'] = NAR_databases_deduplicate['AltName'].replace(r'^\s*$', np.nan, regex=True)
```

```python
# NIHbmic_databases_deduplicates.sample(3)
# list(zip(NIHbmic_databases_deduplicates['Repository_Name'], NIHbmic_databases_deduplicates['Description']))
# NIHbmic_databases_deduplicates.sample(3)
```

```python
NAR = {databasename.strip():description for databasename,description in list(zip(NAR_databases_deduplicate['ShowName'], NAR_databases_deduplicate['Description']))}
NAR_url = {databasename.strip():description for databasename,description in list(zip(NAR_databases_deduplicate['ShowName'], NAR_databases_deduplicate['DBLink']))}
NAR['Cambridge Structural Database'] = 'Crystal structure information for organic and metal organic compounds.'
NAR_url['Cambridge Structural Database'] = 'http://www.ccdc.cam.ac.uk/solutions/csd-system/components/csd/'
EBI = {databasename.strip():description for databasename,description in list(zip(EBI_databases_deduplicates['data_resource'], EBI_databases_deduplicates['profile']))}
EBI_url = {databasename.strip():description for databasename,description in list(zip(EBI_databases_deduplicates['data_resource'], EBI_databases_deduplicates['db_link']))}
NIH = {databasename.strip():description for databasename,description in list(zip(NIHbmic_databases_deduplicates['Repository_Name'], NIHbmic_databases_deduplicates['Description']))}
NIH_url = {databasename.strip():description for databasename,description in list(zip(NIHbmic_databases_deduplicates['Repository_Name'], NIHbmic_databases_deduplicates['Repository_Url']))}
NIH['UniProt'] = 'The Universal Protein Resource (UniProt) is a comprehensive resource for protein sequence and annotation data.'
NIH_url['UniProt'] = 'https://www.uniprot.org/'
```

```python
def genDescription(row):
    if row['Source'] == 'NAR':
        return NAR.get(row['dataset'], 'Unknown')
    if row['Source'] == 'NIH':
        return NIH.get(row['dataset'], 'Unknown')
    if row['Source'] == 'EBI':
        return EBI.get(row['dataset'], 'Unknown')
# Apply the function to each row and create a new column 'Dataset_Normalized'
pmid_dataset['Description'] = pmid_dataset.apply(lambda row: genDescription(row), axis=1)

def genUrl(row):
    if row['Source'] == 'NAR':
        return NAR_url.get(row['dataset'], 'Unknown')
    if row['Source'] == 'NIH':
        return NIH_url.get(row['dataset'], 'Unknown')
    if row['Source'] == 'EBI':
        return EBI_url.get(row['dataset'], 'Unknown')
# Apply the function to each row and create a new column 'Dataset_Normalized'
pmid_dataset['url'] = pmid_dataset.apply(lambda row: genUrl(row), axis=1)
```
